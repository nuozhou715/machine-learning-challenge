```python
import pandas as pd
```

# Read the CSV and Perform Basic Data Cleaning

```python
df = pd.read_csv("exoplanet_data.csv")
# Drop the null columns where all values are null
df = df.dropna(axis='columns', how='all')
# Drop the null rows
df = df.dropna()
df.head()
```

# Select your features (columns)

```python
# Set features. This will also be used as your x values.
# Only drop those that will cause colinearity issues, as we do not know which variable can be useful. Just throw everything
# into the algorithms
X = df.drop(['koi_period_err2', 'koi_time0bk_err2', 'koi_depth_err2', 'koi_disposition'], axis = 1)
print(X.shape)
```

# Create a Train Test Split

Use `koi_disposition` for the y values

```python
y = df['koi_disposition']
```

# Pre-processing

Scale the data using the MinMaxScaler and perform some feature selection

```python
# Scale the data
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, MinMaxScaler

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1, stratify=y)
X_scaler = MinMaxScaler().fit(X_train)
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)


# Label-encode dependent variable
label_encoder = LabelEncoder()
label_encoder.fit(y_train)
encoded_y_train = label_encoder.transform(y_train)
encoded_y_test = label_encoder.transform(y_test)
```

```python
for label, original_class in zip(encoded_y_train, y_train):
    print('Original Class: ' + str(original_class))
    print('Encoded Label: ' + str(label))
    print('-' * 12)
```

# Train the Model



```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier
```

```python
classifier.fit(X_train_scaled, y_train)
```

```python
# Print the model accuracy
print(f"Training Data Score: {classifier.score(X_train_scaled, y_train)}")
print(f"Testing Data Score: {classifier.score(X_test_scaled, y_test)}")
```

# Hyperparameter Tuning

Use `GridSearchCV` to tune the model's parameters

```python
from sklearn.model_selection import GridSearchCV
import numpy as np

# Create regularization penalty space
penalty = ['l1', 'l2']

# Create regularization hyperparameter space
C = np.logspace(0, 1, 2)

# Create hyperparameter options
hyperparameters = dict(C=C, penalty=penalty)
```

```python
# Create grid search using 3-fold cross validation
clf = GridSearchCV(classifier, hyperparameters, cv=3, n_jobs=-1)
```

```python
# Train the model with GridSearch
best_model = clf.fit(X_train_scaled, y_train)
```

```python
print(best_model.best_params_)
print(best_model.best_score_)
```

# Test the Accuracy of the Tuned Model on the Test Data

```python
# Fit the tuned model 
classifier = LogisticRegression(C = 10, penalty = 'l1')
classifier.fit(X_test_scaled, y_test)
```

```python
# Print the model accuracy
print(f"Testing Data Score: {classifier.score(X_test_scaled, y_test)}")
```

### Based on the testing data score, the result is very good but just slightly off comparing to the neural network model.
